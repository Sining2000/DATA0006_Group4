```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
price_and_demand = pd.read_csv("price_and_demand.csv")
price_and_demand
```

```python
# Check datatype of the dataframe
price_and_demand.dtypes
```

```python
# check if there is any mising data
price_and_demand.isna().sum()
```

```python
# Convert "SETTLEMENTDATE" column to datetime
price_and_demand["SETTLEMENTDATE"] = pd.to_datetime(price_and_demand["SETTLEMENTDATE"], format="%d/%m/%Y %H:%M")
```

```python
# Extract date from "SETTLEMENTDATE" and create a new 'Date' column
price_and_demand['Date'] = price_and_demand['SETTLEMENTDATE'].dt.date
```

```python
price_and_demand.sample(20)
```

```python
# Calculate maximum daily energy demand and average rrp
max_edemand = price_and_demand.groupby("Date")['TOTALDEMAND'].max()
mean_rrp = price_and_demand.groupby("Date")['RRP'].mean().round(2)
```

```python
mean_rrp
```

```python
weather = pd.read_csv("weather.csv", encoding='latin-1')
weather
```

```python
weather.dtypes
```

```python
weather.isna().sum()
```

```python
missing_value = weather[["Minimum temperature (Â°C)", "Maximum temperature (Â°C)", "Rainfall (mm)", "Speed of maximum wind gust (km/h)", "3pm relative humidity (%)", "3pm MSL pressure (hPa)"]] .isnull()
missing_value
```

```python
weather["Date"] = pd.to_datetime(weather['Date'], format="%d/%m/%Y")
weather.head()
```

```python
# Extract selected variable columns to form new dataframe
selected_variables = ["Date", "Minimum temperature (Â°C)", "Maximum temperature (Â°C)", "Rainfall (mm)", "Speed of maximum wind gust (km/h)", "9am wind speed (km/h)", "9am MSL pressure (hPa)", "3pm relative humidity (%)", "3pm MSL pressure (hPa)"]
processed_data = weather[selected_variables].set_index('Date')
```

```python
processed_data.head()
```

```python
#substitude missing data with the average value of the respective column
processed_data[["Maximum temperature (Â°C)", "Speed of maximum wind gust (km/h)", "3pm relative humidity (%)", "3pm MSL pressure (hPa)"]] = processed_data[["Maximum temperature (Â°C)", "Speed of maximum wind gust (km/h)", "3pm relative humidity (%)", "3pm MSL pressure (hPa)"]].fillna(processed_data[["Maximum temperature (Â°C)", "Speed of maximum wind gust (km/h)", "3pm relative humidity (%)", "3pm MSL pressure (hPa)"]].mean().round(2))
processed_data.tail()
```

```python
processed_data["Average RRP"] = mean_rrp
processed_data["Maximum Energry Demand"] = max_edemand
processed_data.tail()
```
